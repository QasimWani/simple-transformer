```python
import pandas as pd
import numpy as np

# Create a deliberately buggy dataset
data = {
    "id": [1, 2, 3, 3, 4, 5, 6, 7, 8, 9],
    "age": [25, -5, 42, 42, None, 999, 30, 28, None, 40],
    "income": [50000, 60000, None, None, 45000, 7000000, 52000, None, 48000, 51000],
    "city": ["New York", "new york", None, "SF", "SF", "SF", "Los Angeles", "Los Angeles", "Paris", "Paris"],
    "text": [
        " Great product!! ",
        "Terrible service...",
        "okay experience",
        "okay experience",
        "LOVED IT!!!",
        "bad",
        " Meh ",
        None,
        "good, but pricey",
        "terrible service...",
    ],
    "label": ["positive", "negative", "neutral", "neutral", "positive", None, "neutral", "positive", "positive", "negative"],
}

df_buggy = pd.DataFrame(data)
```

```python
# Okay, so clearly there are some samples that are none.
# Some values don't make sense. Like age being less than 0.

df_buggy = df_buggy.dropna(subset=['income', 'city']) # Let's say we remove all observations with no income and city

median_age = df_buggy[df_buggy['age'] > 0]['age'].median()
df_buggy.loc[(df_buggy['age'] < 0) | (df_buggy['age'].isna())] = median_age
```
